```python
import pygame
pygame.init()
size = [400, 900]
screen = pygame.display.set_mode(size)
title = "Pygame Ex"
pygame.display.set_caption(title)
clock = pygame.time.Clock()

left_move = False
right_move = False

color1 = (242, 167, 167)
color2 = (164, 217, 242)
black_color = (0, 0, 0)
white_color = (255, 255, 255)

class Img_object:
    def __init__(self):
        self.x = 0
        self.y = 0
        
    def add_Img(self, address):
        if address[-3:] == 'png':
           self.img = pygame.image.load(address).convert_alpha()
        else:
            self.img = pygame.image.load(address)
            
    def change_size(self, width, height):
        self.img = pygame.transform.scale(self.img,(width, height))
        self.width, self.height = self.img.get_size()

    def show_Img(self):
        screen.blit(self.img,(self.x, self.y))

hero = Img_object()
hero.add_Img("C:/Users/201/Pictures/pyGame/hero.png")
hero.change_size(80,80)
hero.x = round(size[0] / 2) - round(hero.width / 2) 
#캐릭터 중심에 오게 하기 위해 캐릭터 가로 길이의 절반만큼 왼쪽으로 다시 빼어주어야 한다.
hero.y = size[1] - hero.height - 50
hero.move = 5

# hero = pygame.image.load("C:/Users/201/Pictures/pyGame/hero.png").convert_alpha()
# hero = pygame.transform.scale(hero,(80, 80))
# hero_width, hero_height = hero.get_size()
# hero_x = round(size[0]/2)- round(hero_width / 2)
# hero_y = size[1] -hero_height -50


#메인이벤트
system_exit = 0
while(system_exit == 0):
    clock.tick(60)
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            system_exit = 1
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_LEFT:
               left_move = True
            if event.key == pygame.K_RIGHT:
                right_move = True
        elif event.type == pygame.KEYUP:
            if event.key == pygame.K_LEFT:
                left_move = False
            if event.key == pygame.K_RIGHT:
                right_move = False
    if left_move == True:
           hero.x -= hero.move
           if hero.x <= 0:
               hero.x = 0
    elif right_move == True:
            hero.x += hero.move
            if hero.x >= size[0] - hero.width:
                hero.x = size[0] - hero.width
    
    # 입력에 따른, 시간에 따른 변화
    
    screen.fill(black_color)
    hero.show_Img()
    pygame.display.flip()
            
pygame.quit()
```
