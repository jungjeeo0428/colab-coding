```python
import pygame
pygame.init()
size = [400, 900]
screen = pygame.display.set_mode(size)
title = "Pygame Ex"
pygame.display.set_caption(title)
clock = pygame.time.Clock()
color1 = (242, 167, 167)
color2 = (164, 217, 242)
black_color = (0, 0, 0)
white_color = (255, 255, 255)
hero = pygame.image.load("C:/Users/201/Pictures/pyGame/hero").convert_alpha()
#메인이벤트
system_exit = 0
while(system_exit == 0):
    clock.tick(60)
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            system_exit = 1
    print(event)
    # 입력에 따른, 시간에 따른 변화
    
    screen.fill(black_color)
    screen.blit(hero,(0,0))
    
    pygame.display.flip()
            
pygame.quit()

```
