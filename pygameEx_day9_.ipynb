```python
import pygame
pygame.init()
size = [400, 900]
screen = pygame.display.set_mode(size)
title = "Pygame Ex"
pygame.display.set_caption(title)
clock = pygame.time.Clock()

left_move = False
right_move = False
space_on = False


color1 = (242, 167, 167)
color2 = (164, 217, 242)
black_color = (0, 0, 0)
white_color = (255, 255, 255)

class Img_object:
    def __init__(self):
        self.x = 0
        self.y = 0
        
    def add_Img(self, address):
        if address[-3:] == 'png':
           self.img = pygame.image.load(address).convert_alpha()
        else:
            self.img = pygame.image.load(address)
            
    def change_size(self, width, height):
        self.img = pygame.transform.scale(self.img,(width, height))
        self.width, self.height = self.img.get_size()

    def show_Img(self):
        screen.blit(self.img,(self.x, self.y))
#hero
hero = Img_object()
hero.add_Img("C:/Users/201/Pictures/pyGame/hero.png")
hero.change_size(80,80)
hero.x = round(size[0] / 2) - round(hero.width / 2) 
hero.y = size[1] - hero.height - 50
hero.move = 5


#missile 미사일을 스페이스를 누를 때마다 불러와야한다.
missile_list = []   #스페이스를 누룰 때마다 init갔다오면 초기화되면 인덱스 0이다. 위치가 여기여야 새로운 미사일이 인덱스 새로를 다는것.

#메인이벤트
system_exit = 0
while(system_exit == 0):
    clock.tick(60)
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            system_exit = 1
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_LEFT: 
               left_move = True
            if event.key == pygame.K_RIGHT:
                right_move = True
            if event.key == pygame.K_SPACE:
                space_on = True
        elif event.type == pygame.KEYUP:
            if event.key == pygame.K_LEFT:
                left_move = False
            if event.key == pygame.K_RIGHT:
                right_move = False
            if event.key == pygame.K_SPACE: 
                space_on = False

# 입력에 따른, 시간에 따른 변화
    if left_move == True:
           hero.x -= hero.move
           if hero.x <= 0:
               hero.x = 0
    elif right_move == True:
            hero.x += hero.move
            if hero.x >= size[0] - hero.width:
                hero.x = size[0] - hero.width
    
    if space_on == True:
        missile = Img_object()
        missile.add_Img("C:/Users/201/Pictures/pyGame/missile.png")
        missile.change_size(60,60)
        missile.x = hero.x + 10
        missile.y = size[1] - hero.height - missile.height - 50
        missile.move= 6
        
        missile_list.append(missile) #방금전 불러온 이미지 추가 FPS 가 60이여서 1초에 많은 미사일이 만들어진다. 
       
   
    for i in range(-1, -len(missile_list),-1):
        m = missile_list[i]
        m.y -= m.move
        if m.y <= 0 - missile.height:
           print(missile_list)
           del missile_list[i]
           print("========")
           print(missile_list)
           print("========") 
            

            
            
    screen.fill(black_color)
    hero.show_Img()
    for missile in missile_list:
        missile.show_Img()
    
  
   
    pygame.display.flip()
   
pygame.quit()
```
