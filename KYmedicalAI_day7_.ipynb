```python
# 클래스 : 객체를 표현하는 여러가지 기법 중 1

# 대학병원 환자 관리 시스템을 만들자!
# 환자등록, 진료 기록, 약처방 내역, 입원/ 외래...

# 왜 클래스가 필요함???
# 환자가 1명이라면?
name = "김환자"
age = 42
bp = 120

# 환자가 100명이면?
# name1,age1,bp1,name2,age2,bp2...name100,age100,bp100 -> 변수가 너무 많아지는데?

# 1. 클래스란? -> 객체를 만들기 위한 '양식(설계도)'

# 클래스 : 병원에 비치된 '빈 차트 양식' -> 1개
#   객체 : 그 양식으로 찍어낸 '실제 환자 차트' -> 환자 수 만큼

# 클래스의 구성 요소
# - 속성(attribute) : 차트에 적는 칸 -> 이름, 나이, 전화번호, 혈압 (명사), 변수인데 클래스 안쪽에 있는거
# - 기능(method) : 해당 환자한테 하는 행위 -> 처방하다, 입원하다 (동사), 함수인데 클래스 안쪽에 있는거

# 환자 클래스
# 속성 : 이름 나이, 혈압, 진단명...
# 기능 : 접수하다, 처방하다, 퇴원하다...

class Patient:
  def check_in(self):           # 메서드 1 : 접수
    print("환자 접수 완료")
  def discharge(self):          # 메서드 2 : 퇴원
    print("퇴원 처리 완료")

kim = Patient()                 # kim은 Patient 양식(클래스)로 찍어낸 실제 객체 == 인스턴스

kim.check_in()
kim.discharge()
```

```python
# self -> 한 기능에서 다른 기능을 부르고 싶을 때
# 접수(check_in)을 하면 자동으로 차트 출력(print_chart)도 같이 하고싶어!
# 이럴 때 '이 차트 본인'을 가리키는 self 키워드를 쓴다.

class Patient:
  def print_chart(self):
    print("==환자 차트==")
  def check_in(self):
    print("접수 시작")
    self.print_chart() # self = '이 환자 본인' -> 본인의 다른 메서드 호출

kim = Patient()
kim.check_in() # 기능이 실제로 실행되려면 인스턴스가 있어야해, 접수 하면서 차트까지 같이 출력됨
```

```python
# 4. __init__ (생성자, constructor) -> '신규환자 접수' 순간에 자동으로 실행되는 메서드
# 환자를 등록(인스턴스 생성)하는 바로 그 순간 자동으로 호출되는 메서드
# 보통은 여기서 기본 정보를 적어 넣는다.

class Patient:
  def __init__(self):                              # 생성자
    print("__init__ 실행됨 -> 신규 환자 접수!")
    self.greeting = "안녕하세요, 환자분~~~"

  def greet(self):
    print(self.greeting)

kim = Patient() # 이 순간 __init__이 자동 실행됨
lee = Patient() # 환자 한 명 더 등록 -> __init__ 또 실행됨

kim.greet()

# 참고 : 일반 함수랑 비교
def plus(a, b):
  z = 10
  print(a + b)

# def b():
#   print(z)

plus(1, 2)
# b()
# 함수는 내가 직접 불러야 실행되지만
# __init__ 생성자는 인스턴스를 만드는 순간 '자동'으로 실행되는 점이 차이점이다.
```

```python
# 5. 속성 -> 환자마다 다른 정보를 차트에 담기
# 접수할 때 이름과 나이를 받아서 해당 환자의 차트에 적는다.

class Patient:
  def __init__(self, name, age): # 생성자의 매개변수가 여러개
    self.greeting = "안녕하세요"
    self.name = name             # 넘겨받은 값을 해당 변수(속성)에 저장
    self.age = age

  def introduce(self):
    # 클래스 '내부'에서 속성에 접근할 때는 'self.속성' 으로 써야한다
    print("{} 저는 {}살 {}입니다.".format(self.greeting, self.age, self.name))

kim = Patient("김환자", 56) # 이 때 생성자는 실행되기 때문에 매개변수에 맞게 인자 전달
kim.introduce()
print("이름 : {}".format(kim.name)) # 클래스 '외부'에서 접근 -> '인스턴스.속성'으로 써야한다
print("나이 : {}".format(kim.age))

# 6. 여러 환자 -> 차트 하나에 환자 한 명씩(인스턴스가 여러 개)
# 양식(클래스)은 하나지만, 찍어낸 차트(인스턴스)는 서로 독립이다.

kim = Patient("김환자", 56) # 차트를 만들면서 이름, 나이 칸을 채운다.
kim.introduce()
print("이름 : {}".format(kim.name))

lee = Patient("이영희", 33)
lee.introduce()
print("이름 : {}".format(lee.name))

# kim의 정보와 lee의 정보는 절대로 섞이지 않는다. 각자 본인 차트가 있음
```

```python
# 8. 비공개 속성(private) -> 진단명은 함부로 못건드리게

# 환자의 '진단명' 같은 민감 정보는 외부에서 막 바꾸면 안된다.
# 변수 앞에 __를 붙이면 클래스 '내부'에서만 접근 가능한 비공개 속성이 된다.

class Patient:
  def __init__(self, name, diagnosis):
    self.name = name
    self.__diagnosis = diagnosis  # 비공개 속성

  def update_diagnosis(self, new_diagnosis):
    self.__diagnosis = new_diagnosis # 클래스 내부에서의 접근 -> 허용
    print("{} 환자 진단명 변경! -> {}".format(self.name, self.__diagnosis))

kim = Patient("김철수", "고혈압")
kim.update_diagnosis("당뇨") # 정해진 메서드를 통해서만 변경 가능

# print(kim.diagnosis)
# kim.diagnosis = "당뇨" -> 클래스 외부에서는 못바꾸게 해야해
kim.__diagnosis = "감기"
print(kim.__diagnosis)

print(kim.__dict__) # 외부에서 접근이 된게 아니라, 원본에 영향을 주지 않으며 새로 생성된 속성을 활용한 것
```

```python
# 10. 클래스 속성 vs 인스턴스 속성 -> '병원 공용' vs '환자 개인'
# 어떤 정보는 환자 개인 것이고, 어떤 정보는 병원 전체가 공유한다

# 인스턴스 속성 -> 그 환자 개인의 약 목록(사람마다 다름)
# 클래스 속성 -> 병원 전체 입원 환자 명단(모두가 같이 보는 하나)

class Patient:
  ward_list = [] # 클래스 속성 : 입원 환자 명단 -> 병원에 딱 하나

  def admit(self, name):
    Patient.ward_list.append(name) # 클래스 이름으로 접근

kim = Patient()
kim.admit("김철수")

lee = Patient()
lee.admit("이영희")

print(kim.ward_list)
print(lee.ward_list) # kim으로 보나 lee로 보나 같은 이유는 같은 명단을 공유하기 때문.
# 이게 클래스 속성의 특징(함정)

class Patient:
  def __init__(self):
    self.meds = [] # 인스턴스 속성 : 이 환자 개인의 약 처방 목

  def prescribe(self, drug):
    self.meds.append(drug)  # self로 접근해야함 -> 이 환자 것만

kim = Patient()
kim.prescribe("아스피린")

lee = Patient()
lee.prescribe("타이레놀")

print(kim.meds)
print(lee.meds)
# 공유냐 개인이냐
```

```python
# 12. 정적 메서드(staticmethod) -> 환자가 누구든 상관없어
# BMI 계산은 '이 환자' 가 누군지 관심없음. 키, 몸무게만 있으면 됨
# 이렇게 특정 환자 데이터(self)가 필요 없는 기능은 정적 메서드로 만든다.
# -> 첫번째 매개변수에 self가 없어!

class Health:
  @staticmethod
  def bmi(weight, height): # self가 없음
    return round(weight / (height ** 2), 1)

  @staticmethod
  def dose(weight, mg_per_kg): # 체중당 약 용량 계산
    return weight * mg_per_kg

print(Health.bmi(70,1.75)) # 인스턴스를 만들지 않고 클래스에서 바로 호출
print(Health.dose(70,5))

# '환자 한 명을 등록하지 않아도 쓸 수 있는 계산 도구' -> 정적 메서드의 존재 이유
```

```python
# 13. 클래스 메서드(classmethod) -> '현재 입원 환자 수' 세기

# 환자가 등록될 때마다 카운트를 1 올리고, 전체 환자 수를 알려주는 기능
# 이건 특정 환자(self)가 아니라 '클래스 전체'를 다루는 일이라서 cls를 쓴다


class Patient:
  count = 0 # 클래스 속성 : 등록된 환자 수

  def __init__(self):
    Patient.count += 1 # 등록될 때마다 +1

  @classmethod
  def print_count(cls): # 클래스 메서드 (첫 매개변수가 cls)
    print("현재 등록 환자 수 : {}명".format(cls.count))

Patient.print_count()
kim = Patient()
Patient.print_count()
lee = Patient()
Patient.print_count()
```

```python
# 14. 상속(inheritance) -> 환자에서 '입원환자/외래환자'로 갈라지기

# 입원환자, 외래환자는 환자인가? -> o
# 공통 기능은 부모 클래스에 한 번만 만들고,
# 입원환자만의 추가 기능은 자식 클래스에 더한다.
# 부모 클래스의 속성과 기능을 자식 클래스에게 물려준다.

# 물려주는 쪽 = 부모 클래스
# 물려받는 쪽 = 자식 클래스

class Patient:              # 부모 : 모든 환자의 공통 클래스
  def check_in(self):
    print("접수!")
  def check_in2(self):
    print("접수222222!")

class Inpatient(Patient):   # 자식 : 입원환자 -> Patient 를 물려받음
  def assign_room(self):    # 입원환자만의 기능
    print("병실 배정!")

  def check_in2(self):      # 메소드 오버라이드 -> 덮어쓰기
    print("접수33333333333!")

class Outpatient(Patient):  # 자식 : 외래환자 -> Patient 를 물려받음
  def book_next(self):      # 외래환자만의 기능
    print("다음 외래 예약!")

kim = Inpatient()
kim.check_in()      # 부모한테 물려받은 기능
kim.assign_room()   # 자기만의 기능

kim.check_in2()

lee = Outpatient()
lee.check_in()      # 똑같이 부모한테 물려받은 기능
lee.book_next()     # 자기만의 기능
# lee.assign_room() 이건 안됨

lee.check_in2()

# 자식 클래스의 메서드 갯수는 부모보다 많거나 같다 (물려받은거 + 내꺼)
```
