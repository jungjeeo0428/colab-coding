```python
# for문으로 1부터 n사이에 존재하는 소수의 합을 반환하는 함수 구현
def prime_number(j):
  if j == 1:
    return False
  for i in range(2, j):
    if i < j:  #for i in range(2, n)
      if j % i == 0:
        return False
  else:
    return True

def sum_of_prime_number(n):
  s = 0
  for j in range(1, n+1):
    if prime_number(j): # 여기는 n이 아니라 j 이다. 위의 함수도 j 를 넣어야 한다. value 가 헷갈릴 수 있으니 주의
      s += j
  return(s)

print(sum_of_prime_number(10))
```
